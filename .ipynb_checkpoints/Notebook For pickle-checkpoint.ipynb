```python
df = pd.read_csv('df_copy2.csv')
```

```python
df.head()
```

```python
df_pickle = df[['age', 'ClaimAmount' , 'MaritalStatus', 'Occupation', 'Exposure',\
                'IndustryType', 'SumAssured', 'VehicleMake', 'PersonProvince',\
                'Colour', 'BodyType','VehicleType', 'Kilowatts',\
                'VehicleYear', 'PreviousInsurerPremium',\
                'PreviousInsurerExcess','Gender', 'Transmission', 'VehicleAge']]
```

```python
df_pickle.head()
```

```python
# Numerical columns binarization
n_bins = 20
numeric = [i for i in list(df_pickle.columns) if df_pickle.dtypes[i]!='object' 
                if i not in ['ClaimAmount', 'Exposure', 'PolicyId']]

binarizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='uniform')

df_pickle.loc[:, numeric] = binarizer.fit_transform(X=df_pickle[numeric])
```

```python
#Categorical columns hot encoding
category = [i for i in df_pickle.columns if df_pickle.dtypes[i]=='object']
for col in category:
    one_hot = pd.get_dummies(df_pickle[col], prefix=col, prefix_sep='_', drop_first=True)
    df_pickle = df_pickle.join(one_hot)
    df_pickle.drop([col], inplace=True, axis=1)
```

```python
df_pickle.head()
```

## Model

```python
X = df_pickle.drop(["ClaimAmount"], axis=1)
y = df_pickle["ClaimAmount"]
```

```python
# set aside 20% of train and test data for evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y,
    test_size=0.2, shuffle = True, random_state = 8)

# Use the same function above for the validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
    test_size=0.25, random_state= 8) # 0.25 x 0.8 = 0.2

print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_train shape: {}".format(y_train.shape))
print("y_test shape: {}".format(y_test.shape))
print("X_val shape: {}".format(y_train.shape))
print("y val shape: {}".format(y_test.shape))
```

```python
# Scaling Numerical Columns
scaler = MinMaxScaler(feature_range=(0, 1), copy=True, clip=False)

categorical = [i for i in list(df_pickle.columns) if i not in numeric 
                if i not in ['ClaimAmount', 'PolicyId', 'Exposure']]
X_train = pd.concat([pd.DataFrame(scaler.fit_transform(X_train[numeric]), columns=numeric), 
                     X_train[categorical].reset_index(drop=True), X_train["Exposure"].reset_index(drop=True)], axis=1)

X_val = pd.concat([pd.DataFrame(scaler.transform(X_val[numeric]), columns=numeric), 
                     X_val[categorical].reset_index(drop=True), X_val["Exposure"].reset_index(drop=True)], axis=1)

X_test = pd.concat([pd.DataFrame(scaler.transform(X_test[numeric]), columns=numeric), 
                     X_test[categorical].reset_index(drop=True), X_test["Exposure"].reset_index(drop=True)], axis=1)
```

```python
X_train_prep = X_train.drop(["Exposure"], axis=1)
X_val_prep = X_val.drop(["Exposure"], axis=1)
X_test_prep = X_test.drop(["Exposure"], axis=1)
```

## LinearRegression()

```python
lr = LinearRegression()
```

```python
# Train the model on the training set
model = lr.fit(X_train_prep, y_train)

# model.score(X_val_prep, y_val)

# Let's make prediction on the test data
preds = lr.predict(X_test_prep)
```
