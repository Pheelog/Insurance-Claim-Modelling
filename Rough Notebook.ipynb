```python
insurance_data = pd.read_excel('DataSet.xlsx')
pd.set_option('display.max_columns', None)
```

```python
data = insurance_data.copy()
```

```python
data.head()
```

```python
data.isnull().mean().sort_values(ascending=False)*100
```

```python
#Creating new columns

data['PaymentDueday'] = data['PaymentDueDate'].dt.day
data['PaymentDuemonth'] = data['PaymentDueDate'].dt.month
data['PaymentDueyear'] = data['PaymentDueDate'].dt.year

data.head()
```

```python
now = pd.Timestamp('now')
data['BirthDt'] = pd.to_datetime(data['BirthDt'], format = '%m%d%y')
data['BirthDt'] = data['BirthDt'].where(data['BirthDt'] < now, data['BirthDt'] - np.timedelta64(100, 'Y'))
data['Age'] = (now - data['BirthDt']).astype('<m8[Y]')
```

```python
data['month_full'] = data['PaymentDueDate'].dt.month_name().str[:3]
```

```python
data = data.iloc[:,:-1]
data['SumTotalClaimAmount'].fillna(0, inplace=True)
data['CountTotalClaimAmount'].fillna(0, inplace=True)
```

```python
data.fillna(data.mean(), inplace=True)
```
