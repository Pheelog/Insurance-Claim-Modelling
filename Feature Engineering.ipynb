### **Importing Libraries**

```python
import pandas as pd
import numpy as np
```

### **Importing DataSet**

```python
data = pd.read_excel('DataSet.xlsx')
pd.set_option('display.max_columns', None)
```

```python
#creating a copy of the data so as not to alter the original data
df = data.copy()
```

```python
#check the first five rows of the dataset
df.head()
```

```python
#checking the number of rows and columns in the dataset
df.shape
```

### **Droping irrelevant columns**

```python
print(list(df.columns))
```

```python
#droping irrelevant columns
df.drop(['PremiumSum', 'CasePremiumSum', 'PaymentDueDate','TotalPremiumPaymentCount',0.5865768182345752], axis = 1, inplace=True)

df.head()
```

### **Checking for Missing Values**

```python
df.isnull().mean().sort_values(ascending=False)
```

### **Dropping Values with high percentage of missing values**

```python
#dropping values with high percentages of missing values
df.drop(['Employer', 'MonthsSinceInception', 'DaysSinceInception'], axis = 1, inplace=True)

df.head()
```

### **Droping Redundant columns**

```python
df.drop(['Area'], axis = 1, inplace=True)

df.head()
```

```python
df.shape
```

### **Impute missing values**

```python
#Imputing missing values by replace the value for the mean in some numerical columns and the mode in categorical columns
df['SumTotalClaimAmount'].fillna(0, inplace=True)
df['CountTotalClaimAmount'].fillna(0, inplace=True)
df['PreviousInsurerExcess'].fillna(df['PreviousInsurerExcess'].mean(), inplace=True)
df['SumAssured'].fillna(df['SumAssured'].mean(), inplace=True)
df['Occupation'].fillna(df['Occupation'].mode()[0], inplace=True)
df['IndustryType'].fillna(df['Occupation'].mode()[0], inplace=True)
df['EmploymentType'].fillna(df['Occupation'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

### **Droping multiple missing value rows**

```python
df.dropna(axis=0, thresh=18, inplace=True)
df.head()
```

### **Calculating Age Column**

```python
now = pd.Timestamp('now')
df['BirthDt'] = pd.to_datetime(df['BirthDt'], format = '%m%d%y')
df['BirthDt'] = df['BirthDt'].where(df['BirthDt'] < now, df['BirthDt'] - np.timedelta64(100, 'Y'))
df['Age'] = (now - df['BirthDt']).astype('<m8[Y]')
df.head()
```

### **Changing Uppercases to Lowercases in some columns**

```python
df['Occupation'] = df['Occupation'].str.upper()
df.head()
```

### **Extarcting the First name of car model**

```python
df['Model'] = df.Model.str.split().str.get(0)
df.head()
```

```python
category = [i for i in df.columns if df.dtypes[i]=='object']
top_n = 10
for cat in category:
    top_cats  = df[cat].value_counts().head(top_n).index.tolist()
    df.loc[~df[cat].isin(top_cats), cat] = 'Others'
```
