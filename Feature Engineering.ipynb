**Calculating Age Column**

```python
now = pd.Timestamp('now')
df['BirthDt'] = pd.to_datetime(df['BirthDt'], format = '%m%d%y')
df['BirthDt'] = df['BirthDt'].where(df['BirthDt'] < now, df['BirthDt'] - np.timedelta64(100, 'Y'))
df['Age'] = (now - df['BirthDt']).astype('<m8[Y]')
```

```python
df.head()
```

**Changing Uppercases to Lowercases**

```python
df['Occupation'] = df['Occupation'].str.upper()
```

```python
make = df[df['Make']=='TOYOTA'][['Make','Model']]
make.head(20)
```

```python
now = pd.Timestamp('now')
df['CommencementDt'] = pd.to_datetime(df['CommencementDt'], format = '%m%d%y')
df['CommencementDt'] = df['CommencementDt'].where(df['CommencementDt'] < now, df['CommencementDt'] - np.timedelta64(100, 'Y'))
df['Exposure'] = (now - df['CommencementDt']).astype('<m8[Y]')
```

```python
df['ClaimFrequency'] = df['CountTotalClaimAmount']/df['Exposure']
```
